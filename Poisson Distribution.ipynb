```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import exp
from math import factorial as f
import seaborn as sns
%matplotlib inline
```

```python
#the expected value is 1.6, what is the probability of 3 customers?
global lamb,x
lamb = 1.6
def poisson(x):
    res = ((lamb**x)*exp(-lamb))/f(x)
    return res
```

```python
df = pd.DataFrame()
df['outcome'] = np.arange(11)
df['prob'] = df['outcome'].apply(lambda x:poisson(x)).round(6)
df['cum'] = df['prob'].cumsum()
```

```python
plt.figure(figsize=(9,6),dpi=200)
#barplot = plt.bar(df['outcome'],df['prob'])
barplot = sns.barplot(x = 'outcome', y = 'prob', data = df,palette='magma')
for bar in barplot.patches:
    barplot.annotate(format(bar.get_height(), '.3f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=7, xytext=(0, 6),
                   textcoords='offset points')

plt.title('Probability Distribution')
```

```python
print("The probability of getting 3 customers or less is",df['cum'][3].round(3))
```
